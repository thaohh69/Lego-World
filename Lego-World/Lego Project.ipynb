```python
%matplotlib inline

sets = pd.read_csv('https://raw.githubusercontent.com/indianspice/Lego/master/Data/sets.csv')
sets.head()
```

```python
# Create a summary of average number of parts by year: `parts_by_year`

parts_by_year = sets[['year', 'num_parts']].groupby('year', as_index=False).mean().round(2)
parts_by_year
```

```python
# Plot trends in average number of parts by year
import matplotlib.pyplot as plt
plt.plot('year', 'num_parts', data = parts_by_year)
plt.xlabel('Year')
plt.ylabel('Average Number of Parts')
plt.title('Average Lego Parts by Year')
plt.show()
```

```python
# themes_by_year: Number of themes shipped by year
themes_by_year = sets[['year','theme_id']].groupby('year').theme_id.nunique()
themes_by_year
```
